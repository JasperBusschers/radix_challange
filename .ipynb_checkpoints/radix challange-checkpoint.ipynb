### preprocess data

```python
def prepare_image(image_path):
    """
    Prepares an image for processing: resizes width to 224 pixels, crops height to 224 pixels from the center,
    and converts to RGB if necessary.
    :param image_path: Path to the image file.
    :return: Processed Image object.
    """
    image = Image.open(image_path)
    # Resize image maintaining aspect ratio
    base_width = 224
    w_percent = (base_width / float(image.size[0]))
    h_size = int((float(image.size[1]) * float(w_percent)))
    image = image.resize((base_width, h_size), Image.ANTIALIAS)
    # Crop the center of the image
    top = (h_size - 224) // 2
    bottom = top + 224
    image = image.crop((0, top, 224, bottom))
    # Check if image is not RGB (e.g., grayscale) and convert to RGB
    if image.mode != 'RGB':
        image = image.convert('RGB')
    return image
    
def getData(moviesDf,label):
    keys = moviesDf['key']
    genre = moviesDf['genres'].tolist()
    targetGenre = [[0.,1.] if label in genre else [1.,0.] for genre in genre]
    synopsis = moviesDf['synopsis']
    images = []
    for i ,k in enumerate(keys):
        image_path = os.path.join('data' ,f'{k}.png')
        image = prepare_image(image_path)
        images.append(image)
    return images,synopsis,targetGenre

def preprocessLabel(dataframe):
    data=pd.DataFrame({})
    for value in genres:
        data[value] = 0
    import ast
    # Step 3: Update new columns based on list values
    for index, row in dataframe.iterrows():
                
                genre=row['genres']
                if isinstance(genre, str):
                    genre=ast.literal_eval(genre)
                for item in genre:
                    if item in genres:
                        data.at[index, item] = 1
                    else:
                        data.at[index, item] = 0
    return data.fillna(0)
y_train  = preprocessLabel(moviesTrain)
y_val  = preprocessLabel(moviesVal)
y_test  = preprocessLabel(moviesTest)
```

### fine tune language embedding model
Here we will fine tune a language embedding model to minimize distances between movies of the same genre.

```python
from torch.utils.data import DataLoader
from sentence_transformers import SentenceTransformer, SentencesDataset, InputExample, losses
from sentence_transformers.evaluation import TripletEvaluator


model = SentenceTransformer('all-mpnet-base-v2')

from sentence_transformers import InputExample
import ast
import random

# Creating a dictionary to group synopses by genre
genre_dict = {}
for index, row in moviesTrain.iterrows():
    genre=row['genres']
    if isinstance(genre, str):
        genre= ast.literal_eval(genre)
    for g in genre:
        if g not in genre_dict:
            genre_dict[str(g)] = []
        genre_dict[str(g)].append(row['synopsis'])
movies_per_genre=50
def create_triplet_dataset(df):
    anchors = []  # List of anchor genres
    positives = []  # List of positive synopses
    negatives = []  # List of negative synopses
    # Create triplets
    for i in range(1):
        for genre, synopses in genre_dict.items():
            for synopsis in synopses[:min(len(synopses),movies_per_genre)]:
                # Positive example: another synopsis from the same genre
                positive_examples = [s for s in synopses if s != synopsis]
                if positive_examples:
                    positive_example = random.choice(positive_examples)
    
                    # Negative example: a synopsis from a different genre
                    other_genres = [g for g in genre_dict if g != genre]
                    if other_genres:
                        negative_genre = random.choice(other_genres)
                        negative_example = random.choice(genre_dict[negative_genre])
                        # Add to the respective lists
                            
                        anchors.append('The genre of the movie is '+genre)
                        positives.append(positive_example)
                        negatives.append(negative_example)
    
    return anchors, positives, negatives


anchors, positives, negatives=create_triplet_dataset(moviesTrain)
train_examples = []
for anchor, pos, neg in zip(anchors, positives, negatives):
    train_examples.append(InputExample(texts=[anchor, pos, neg]))
train_dataset = SentencesDataset(train_examples, model)
train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=32)
anchors_val, positives_val, negatives_val=create_triplet_dataset(moviesVal)
evaluator = TripletEvaluator(anchors_val, positives_val, negatives_val)


# Define the loss function
train_loss = losses.TripletLoss(model=model)


# Fine-tune the model
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=1, warmup_steps=100, evaluator=evaluator, evaluation_steps=500)


# Save the model
model.save("tuned_models/fine_tuned_model")
```

# Create text features
We use this to embed sentences by taking the distance from the label embedding resulting in 19 features
(another way could have been to cluster the embeddings)<br>
This option has been chosen since it results in interpretable features (Whether a math has been found in the text for genre x)

```python
from torch.nn import CosineSimilarity
def embedSentences(sentences):
    model = SentenceTransformer('tuned_models/fine_tuned_model').to(device)

    embeddingsTargets = torch.Tensor(model.encode(['The genre of the movie is '+g for g in genres])).to(device)#random.choice(genre_dict[g])) for g in genres]).to(device)       
        
    # Sentences we want to encode. Example:
    sentence = sentences
    # Sentences are encoded by calling model.encode()
    embeddings = torch.Tensor(model.encode(sentence)).to(device)
    mse_per_rows=[]
    for e in embeddingsTargets:
        
        # Broadcast y_pred to match the shape of y_true
        e_expanded = e.expand_as(embeddings)
        
        # Calculate the squared differences
        cos = nn.CosineSimilarity(dim=1, eps=1e-6)
        distances = cos(e_expanded, embeddings) 
        
        # Calculate the mean squared error for each row
        #mse_per_row = torch.mean(squared_diff, dim=1)
        mse_per_rows.append(distances.flatten())
    # Concatenating the results and converting to a list
    mse_matrix = torch.stack(mse_per_rows, dim=1)
    output = {}
    for i, genre in enumerate(genres):
        output[genre + "_txt"] = [emb[i] for emb in mse_matrix.tolist() ]
    features=pd.DataFrame(output)
    return features

_,texts,_ = getData(moviesTrain,"")
# Process text embeddings
textEmbeddings = embedSentences(texts.tolist())
df_merged = pd.concat([textEmbeddings, y_train], axis=1, sort=False)


sns.set(font_scale=0.5)
plt.figure(figsize=(16, 12))
sns.heatmap(df_merged.corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)
plt.show()

```

## fine tune imagenet model to extract image features
19 binary classifiers are trained to predict each genre, the predictions are concatenated to result in 19 features related to the image.

```python
os.environ["WANDB_DISABLED"] = "true"
#fine tune imagenet model from huggingface
def fineTuneImagenet(x,y,x_val,y_val,label):
    y_labels=np.argmax(y,axis=1)
    class_weights = compute_class_weight('balanced', classes=np.unique([0,1]), y=y_labels)
    class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

    # Update the CustomTrainer class to use the newly calculated class_weights
    class CustomTrainer(Trainer):
        def compute_loss(self, model, inputs, return_outputs=False):
            labels = inputs.pop("labels")
            outputs = model(**inputs)
            loss_fct = CrossEntropyLoss(weight=class_weights)
            loss = loss_fct(outputs.logits, labels)
            return (loss, outputs) if return_outputs else loss
    model_name='google/vit-base-patch16-224'
    processor = ViTImageProcessor.from_pretrained(model_name)
    data_dict = {"img":x ,"label":y}
    # Create dataset
    trainds = Dataset.from_dict(data_dict)
    data_dict_val = {"img":x_val ,"label":y_val}
    valds = Dataset.from_dict(data_dict_val)
    mu ,sigma = processor.image_mean ,processor.image_std
    norm = Normalize(mean=mu ,std=sigma)
    size = processor.size
    _transf = Compose([
        Resize(size['height']) ,
        ToTensor() ,
        norm
    ])
    def transf(arg):
        arg['pixels'] = [_transf(image.convert('RGB')) for image in arg['img']]
        return arg
    trainds.set_transform(transf)
    valds.set_transform(transf)
    model = ViTForImageClassification.from_pretrained(model_name ,num_labels=2 ,
                                                      ignore_mismatched_sizes=True )

    args = TrainingArguments(
        run_name='radix'   ,
        output_dir='topic_classification' ,
        save_strategy="steps" ,
        evaluation_strategy="steps" ,
        learning_rate=5e-5 ,
        per_device_train_batch_size=64 ,
        per_device_eval_batch_size=10 ,
        num_train_epochs=1 ,
        weight_decay=0.01 ,
        load_best_model_at_end=True ,
        do_train=True ,
        do_eval=True ,
        metric_for_best_model="f1" ,
        remove_unused_columns=False,
        logging_steps=2,
        eval_steps=2,
        report_to=None

    )

    def collate_fn(examples):
        pixels = torch.stack([example["pixels"] for example in examples])
        labels = torch.tensor([example["label"] for example in examples])
        return {"pixel_values":pixels ,"labels":labels}

    def compute_metrics(eval_pred):
        predictions ,labels = eval_pred
        return dict(accuracy=accuracy_score(np.argmax(predictions,axis=1) ,np.argmax(labels,axis=1)),precision=precision_score(np.argmax(predictions,axis=1) ,np.argmax(labels,axis=1)),recall=recall_score(np.argmax(predictions,axis=1) ,np.argmax(labels,axis=1)))

    trainer = CustomTrainer (
        model ,
        args ,
        train_dataset=trainds ,
        eval_dataset=valds ,
        data_collator=collate_fn ,
        compute_metrics=compute_metrics ,
        tokenizer=processor ,
    )
    trainer.train()
    trainer.save_model("radix_"+label)




```

## Setup evaluation metrices

```python
import numpy as np

def apk(actual, predicted, k=10):
    """
    Computes the average precision at k.

    This function computes the average prescision at k between two lists of
    items.

    Parameters
    ----------
    actual : list
             A list of elements that are to be predicted (order doesn't matter)
    predicted : list
                A list of predicted elements (order does matter)
    k : int, optional
        The maximum number of predicted elements

    Returns
    -------
    score : double
            The average precision at k over the input lists

    """
    if len(predicted)>k:
        predicted = predicted[:k]

    score = 0.0
    num_hits = 0.0

    for i,p in enumerate(predicted):
        if p in actual and p not in predicted[:i]:
            num_hits += 1.0
            score += num_hits / (i+1.0)

    if not actual:
        return 0.0

    return score / min(len(actual), k)

def mapk(actual, predicted, k=10):
    """
    Computes the mean average precision at k.

    This function computes the mean average prescision at k between two lists
    of lists of items.

    Parameters
    ----------
    actual : list
             A list of lists of elements that are to be predicted 
             (order doesn't matter in the lists)
    predicted : list
                A list of lists of predicted elements
                (order matters in the lists)
    k : int, optional
        The maximum number of predicted elements

    Returns
    -------
    score : double
            The mean average precision at k over the input lists

    """
    return np.mean([apk(a,p,k) for a,p in zip(actual, predicted)])
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
import pandas as pd
import matplotlib.pyplot as plt

def train_decision_tree(X_dataframe, y_dataframe):
    model = DecisionTreeClassifier(class_weight='balanced',max_depth=4)
    model.fit(X_dataframe, y_dataframe)
    return model

def explain_decision_tree(model, feature_names, class_names):
    # Textual Explanation
    tree_text = export_text(model, feature_names=feature_names)
    print("Text Representation of the Decision Tree:\n")
    print(tree_text)

from sklearn.metrics import classification_report, accuracy_score

def evaluate_decision_tree(model, X_test, y_test):
    # Predictions from the model
    y_pred = model.predict(X_test)

    # Calculate and print the accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.2f}")

    # Detailed classification report
    report = classification_report(y_test, y_pred)
    print("\nClassification Report:\n")
    print(report)
     # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt="d")
    plt.title('Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()

```

## Neural network model for multi-class classifier instead of multiple binary models

```python
class NeuralNetwork:
    """ A modular neural network class for easy customization and extension. """

    def __init__(self, input_size, output_size, hidden_size=10, depth=1):
        self.input_size = input_size
        self.output_size = output_size
        self.hidden_size = hidden_size
        self.depth = depth
        self.model = self._build_model().to(device)

    def _build_model(self):
        # Add the initial linear layer and batch normalization layer
        layers = [nn.Linear(self.input_size, self.hidden_size), nn.BatchNorm1d(self.hidden_size), nn.ReLU()]
        # Add additional layers
        for _ in range(self.depth - 1):
            layers.extend([nn.Linear(self.hidden_size, self.hidden_size), nn.BatchNorm1d(self.hidden_size), nn.ReLU()])
        # Add the final layer
        layers.extend([nn.Linear(self.hidden_size, self.output_size), nn.Sigmoid()])
        return nn.Sequential(*layers)


    def set_weights(self, weights):
        start = 0
        for layer in self.model:
            if isinstance(layer, nn.Linear):
                for param in [layer.weight, layer.bias]:
                    end = start + param.numel()
                    param.data = weights[start:end].view(param.size())
                    start = end

    def evaluate(self, x, y):
        pred = self.model(x)
        loss = nn.BCELoss()
        return loss(pred, y).item()

    def get_solution_length(self):
        return sum(param.numel() for param in self.model.parameters())
```

## Find best weight initialization using evolutionary algorithms

```python
from evotorch.logging import StdOutLogger
from evotorch.algorithms import SteadyStateGA
from evotorch.algorithms import SNES
from torchmetrics import Precision
from torch.nn import MSELoss
x_train_tensor = torch.tensor(np.asarray(xTrain.values), dtype=torch.float32).to(device)
y_train_tensor = torch.tensor(np.asarray(y_train.values), dtype=torch.float32).to(device)
x_val_tensor = torch.tensor(np.asarray(xVal.values), dtype=torch.float32).to(device)
y_val_tensor = torch.tensor(np.asarray(y_val.values), dtype=torch.float32).to(device)
x_test_tensor = torch.tensor(np.asarray(xTest.values), dtype=torch.float32).to(device)
y_test_tensor = torch.tensor(np.asarray(y_test.values), dtype=torch.float32).to(device)

def fitness_function(weights, returnBest=False):
    best_score =0
    best_weights = None
    # Initialize the neural network with the given weights
    # Convert data to torch tensors (assuming x_train, y_train, etc. are defined)

    bce_scores = []
    mapk_scores = []
    validation_scores=[]
    test_scores=[]
    for w in tqdm.tqdm(weights):
        net = NeuralNetwork(38, 19)
        weight=w.clone()
        net.set_weights(weight.to(device))
        # Training process (simplified for a few epochs)
        optimizer = Adam(net.model.parameters(), lr=5e-4)
        for epoch in range(10):  # Train for a small number of epochs
            optimizer.zero_grad()
            y_pred = net.model(x_train_tensor)
            loss = nn.MSELoss()(y_pred, y_train_tensor)
            loss.backward()
            optimizer.step()
        ranked_predictions = [[j for j, p in enumerate(pred.cpu().detach().numpy().tolist()) if p>0.5] for pred in y_pred]
        actual = [list(np.where(label.cpu().numpy() == 1)[0]) for label in y_train_tensor]
        k= 10  # Choose an appropriate k value
        mapk_score = mapk(actual, ranked_predictions, k)
        mapk_scores.append(mapk_score)
            
    
        # Evaluate the model on the validation set
        with torch.no_grad():
            y_val_pred = net.model(x_val_tensor)
            # Other evaluation metrics can be included here
            ranked_predictions = [[j for j, p in enumerate(pred.cpu().numpy().tolist()) if p>0.5] for pred in y_val_pred]
            actual = [list(np.where(label.cpu().numpy() == 1)[0]) for label in y_val_tensor]
            k = 10  # Choose an appropriate k value
            mapk_score = mapk(actual, ranked_predictions, k)
            if mapk_score>best_score:
                best_weights = weight
            validation_scores.append(mapk_score)


            #test
            y_test_pred = net.model(x_test_tensor)
            # Other evaluation metrics can be included here
            ranked_predictions = [[j for j, p in enumerate(pred.cpu().numpy().tolist()) if p>0.5] for pred in y_test_pred]
            actual = [list(np.where(label.cpu().numpy() == 1)[0]) for label in y_test_tensor]
            k = 10  # Choose an appropriate k value
            mapk_score = mapk(actual, ranked_predictions, k)
            test_scores.append(mapk_score)
    print("best MAPK  on val set : ",np.max(validation_scores))
    print("best MAPK  on test set : ",np.max(test_scores))
    if returnBest:
        return best_weights
    return  torch.stack([torch.tensor(mapk_scores) ], dim=-1)  # or another metric of your choice

# Define the problem for the genetic algorithm
problem = Problem(
    "max",  
    fitness_function,
    initial_bounds=(-1, 1),  # Bounds for weight initialization
    solution_length=NeuralNetwork(38, 19).get_solution_length(),  # Length of the weight vector
    vectorized=True,
    device=device,

)

# Initialize the genetic algorithm
ga = SNES(
    problem,
    popsize=100,
    stdev_init=0.9,
)
#ga.use(SimulatedBinaryCrossOver(problem, tournament_size=5, cross_over_rate=0.7, eta=0.1))
#ga.use(GaussianMutation(problem, stdev=0.9))
logger = StdOutLogger(ga)
ga.run(100)
```
